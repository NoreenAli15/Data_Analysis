```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("clean_dataset.csv")
```

```python
df
```

```python
df.info()
```

```python
#pie chart
mar=(df['Married']==1).sum()
notmar=(df['Married']==0).sum()
sizes=[mar,notmar]
labels=['Married','Not Married']
plt.pie(sizes,labels=labels,shadow=True,autopct='%1.1f%%')
plt.legend()
plt.figure(figsize=(66, 66))
plt.savefig('image')
plt.show()
```

```python
#bar chart
import matplotlib.pyplot as plt

approved = (df['Approved'] == 1).sum()
notapproved = (df['Approved'] == 0).sum()

categories = ['Approved', 'Not Approved']
counts = [approved, notapproved]

plt.bar(categories, counts,color = "#4CAF50",width = .66)
plt.xlabel('Approval Status')
plt.ylabel('Count')
plt.title('Count of Approved vs Not Approved')
plt.grid()
plt.show()
```

```python
#histogram
plt.hist(df['YearsEmployed'])
plt.show()
```

```python
#boxplot
sns.boxplot(y="Age", x="Approved", hue='Approved',data=df,palette='viridis')
```

```python
#violinplot
sns.violinplot(x="Married", y="Debt", data=df,hue='Married')
```

```python
#countblot
sns.countplot(x='Married',data=df,palette='rocket',hue='Married')
```

```python
#heatmap
sns.heatmap(df.corr(),annot=True)
```

```python
#distplot
sns.distplot(df['Age'],kde=True)
```

```python
#kdeplot
sns.kdeplot(df['Age'],shade=True)
```

```python
#pairplot
sns.pairplot(df)
```
