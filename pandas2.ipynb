```python
import pandas as pd
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Sam','Sarah'],
       'Sales':[200,120,340,124,243,350],
       'Calls': [60,190,2,100,700,423]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#groupby
df.groupby('Company').mean()
```

```python
df.groupby('Sales').min()
```

```python
df.groupby('Company').describe()
```

```python
df.groupby('Company').describe().transpose()
```

```python
data2 = {'Company':['Foo','GOO','Temp','Conc','Tel','Tel'],
       'Person':['Ahmed','Kevin','Heaven','Marry','Liza','Paula'],
       'Sales':[500,189,387,492,987,750],
       'Calls': [80,20,9,100,170,90]}
```

```python
df2= pd.DataFrame(data2)
df2
```

```python
#concatenation
pd.concat([df,df2])
```

```python
#merging
pd.merge(df,df2,how='outer',on='Calls')
pd.merge(df,df2,how='inner',on='Calls')
pd.merge(df,df2,how='right',on='Calls')
pd.merge(df,df2,how='left',on='Calls')
```

```python
#pivot
pd.pivot_table(df, index=['Company'], columns='Sales', values='Calls')
```
